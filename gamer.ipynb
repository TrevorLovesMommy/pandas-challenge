```python
import pandas as pd
```

```python
file = "02-Homework_04-Pandas_Instructions_HeroesOfPymoli_Resources_purchase_data.csv"
df = pd.read_csv(file)
df.head()
```

Player Count

```python
#sum number of players
num_players = len(df["SN"].unique())

#create summary dataframe with a dictionary
#add index=[0] if using all scalar 
summary_table = pd.DataFrame({"Total":num_players}, index = [0])
#Display totla number of playersd
summary_table
```

Purchasing Analysis (Total)

```python
#Avg Total Purchase per Person
#Number of Unique Items
num_unique_items = len(df["Item ID"].unique())

#Average purchase price
avg_purchase_price = df["Price"].mean()

#Total number of purchases
num_purchases = df["Purchase ID"].count()

#Total revenue
revenue = df["Price"].sum()

#create summary dataframe with a dictionary
purchasing_summary_df = pd.DataFrame({"Number of Unique Items": num_unique_items,
                                     "Average Price":avg_purchase_price,
                                     "Number of Purchases": num_purchases,
                                     "Total Revenue":revenue}, index = [0])

#Optional: give the displayed data cleaner formatting
#avg_purchase_price = avg_purchase_price.map("{:.2%}".format)

#Display the summary data frame
purchasing_summary_df
```

Gender Demographics

```python
#get number of players
population = len(df["SN"].unique())

#reduce dataframe to Sn and Gender
gender_df = df[["SN", "Gender"]]

#filter for unique SN
#gender_df = df.iloc[df["SN"].nunique(),:]
gender_df = df.drop_duplicates(["SN"])

#create dataframe of value counts
gender_df = pd.DataFrame(gender_df["Gender"].value_counts())

#rename first column
gender_df.columns =["Total Count"]


#add column of percentage of players
gender_df["Percentage of Players"]=gender_df["Total Count"]/population

#format percentage of players
gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map("{:.2%}".format)
gender_df.head()
```

 Purchasing Analysis (Gender)


```python
# Using GroupBy in order to separate the data into fields according to "state" values
grouped_df = df.groupby(["Gender"])

# In order to be visualized, a data function must be used...
#grouped_df.count().head(10)

#Purchase Count
purchase_count = grouped_df["Purchase ID"].count()

#Average purchase price
average_purchase_price = grouped_df["Price"].mean()

#Total purchase value
total_purchase_price = grouped_df["Price"].sum()

#Average Total Purchase per person
#----------------------------------
avg_tot_purch_pp= total_purchase_price/gender_df["Total Count"]

#optional - format currency
#avg_tot_purch_pp = avg_tot_purch_pp.map("${:.2f}".format)
total_purchase_price = total_purchase_price.map("${:.2f}".format)
average_purchase_price = average_purchase_price.map("${:.2f}".format)
avg_tot_purch_pp=avg_tot_purch_pp.map("${:.2f}".format)

# Create a summary dataframe with dictionaries
purchase_summary_df = pd.DataFrame({"Purchase Count": purchase_count,
                                   "Average Purchase Price": average_purchase_price,
                                   "Total Purchase Price":total_purchase_price,
                                   "Avg Total Purchase per Person":avg_tot_purch_pp})

#purchase_summary_df.head()
purchase_summary_df.head()

```

Age Demographics

```python
#use unique tables

#reduce dataframe to Sn and Gender
age_demo_df = df[["SN", "Gender"]]

#filter for unique SN
age_demo_df = df.drop_duplicates(["SN"])


#Establish bins for ages
max_age = age_demo_df["Age"].max()
bins = [0, 9, 14, 19, 24, 29, 34, 39, max_age]
group_names = ["<10", "10-14", "15-19", "20-24","25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins. Hint: use pd.cut()
age_demo_df["Age Group"] = pd.cut(age_demo_df["Age"], bins, labels = group_names)

#group by age group

age_grouped_df = age_demo_df.groupby(["Age Group"])
#age_grouped_df.count().head(10)

#Calculate the numbers and percentages by age group
num_by_bin = age_grouped_df["SN"].count()
percentage_of_players = age_grouped_df["SN"].count()/population

#Optional: round the percentage column to two decimal points
percentage_of_players = percentage_of_players.map("{:.2%}".format)

#Create a summary data frame to hold the results
age_summary_df = pd.DataFrame({"Total count": num_by_bin, 
                            "Percentage of Players": percentage_of_players})

#Display Age Demographics Table
age_summary_df.head(10)
```

Purchasing Analysis (Age)

```python
#Bin the purchase_data data frame by age
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
age_demo_df1 = df[["SN", "Gender","Age", "Purchase ID", "Price"]]

#Establish bins for ages
max_age = age_demo_df1["Age"].max()
bins = [0, 9, 14, 19, 24, 29, 34, 39, max_age]
group_names = ["<10", "10-14", "15-19", "20-24","25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins. Hint: use pd.cut()
age_demo_df1["Age Group"] = pd.cut(age_demo_df1["Age"], bins, labels = group_names)
#group by age group

age_grouped_df1 = age_demo_df1.groupby(["Age Group"])
#age_grouped_df.count().head(10)


purchase_count_age = age_grouped_df1["Purchase ID"]
purchase_count_age = purchase_count_age.count()

avg_purchase_price_age = age_grouped_df1["Price"]
avg_purchase_price_age = avg_purchase_price_age.mean()

total_purchase_price_age = age_grouped_df1["Price"]
total_purchase_price_age = total_purchase_price_age.sum()

ave_purch_price_per_person_age= total_purchase_price_age/num_by_bin

#Optional: give the displayed data cleaner formatting
avg_purchase_price_age = avg_purchase_price_age.map("${:.2f}".format)
total_purchase_price_age = total_purchase_price_age.map("${:.2f}".format)
ave_purch_price_per_person_age = ave_purch_price_per_person_age.map("${:.2f}".format)

#Create a summary data frame to hold the results
purchase_age_summary_df = pd.DataFrame({"Purchase Count": purchase_count_age, 
                                        "Average Purchase Price": avg_purchase_price_age,
                                       "Total Purchase Price":total_purchase_price_age,
                                       "Avg Total Purchase per Person":ave_purch_price_per_person_age})


#Display the summary data frame
purchase_age_summary_df
#age_grouped_df.count().head(10)
```

Post Popular Items

```python
#grouping by Item Name and ID and getting individual values with counta and sum
#code from session with TA
count_pop_df = df.groupby(["Item ID", "Item Name"]).count()
mean_pop_df = df.groupby(["Item ID", "Item Name"]).mean()
sum_pop_df = df.groupby(["Item ID", "Item Name"]).sum()

item_price = mean_pop_df["Price"]
total_price = sum_pop_df["Price"]
purchase_count = count_pop_df["Purchase ID"]

#Optional: give the displayed data cleaner formatting
item_price = item_price.map("${:.2f}".format)
total_price = total_price.map("${:.2f}".format)


data = {"Purchase Count": purchase_count,
       "Item Price": item_price,
       "Total Purchase Value":  total_price}

pop_summary_df = pd.DataFrame(data).sort_values(["Purchase Count"], ascending = False)

pop_summary_df.head()
```
